```python
import pandas as pd

# Read the RTI-final.csv file
rti_final_df = pd.read_csv('RTI-final.csv')

# Read the SA-zones.csv file
sa_zones_df = pd.read_csv('SA-zones.csv')

# Merge zone attributes to RTi-final based on SA2_CODE21 variable (0_ZID in SA-zones file)
df = pd.merge(rti_final_df, sa_zones_df, how='left', left_on='SA2_CODE21', right_on='0_ZID')

# Display the merged dataframe
print(df)
```

```python
def analyze_data(df):
    """
    Print column names, top 2 most frequent values from each column, and type of each column.

    Parameters:
    df (pandas.DataFrame): The data frame to analyze.
    """
    for column in df.columns:
        print(f"Column: {column}")
        print(f"Type: {df[column].dtype}")
        print("Top 20 most frequent values:")
        print(df[column].value_counts().head(20))
        print("\n")

analyze_data(df)
```

```python
import pandas as pd

# Step 1: Convert string representation of lists into actual lists
# Replace empty string lists with a placeholder for easier handling
df['Attending_Groups'] = df['Attending_Groups'].replace('unknown', '[]')

df['Attending_Groups'] = df['Attending_Groups'].replace("[' ']", "['Other']").apply(eval)

# Step 2: Identify unique groups and aggregate less frequent ones if necessary
all_groups = set(group for sublist in df['Attending_Groups'] for group in sublist)

# Step 3: Create binary columns for each unique group
for group in all_groups:
    # Create a new column for each group, marking 1 if the group is present, 0 otherwise
    df[group] = df['Attending_Groups'].apply(lambda x: 1 if group in x else 0)
```

```python
from sklearn.preprocessing import LabelEncoder
import re

def extract_info_from_display_name(row):
    incident_type = 'unknown'
    num_vehicles_involved = 1  # Default to 1 if specific number is not mentioned

    if 'breakdown' in row.lower():
        incident_type = 'breakdown'
    elif 'crash' in row.lower():
        incident_type = 'crash'
    elif 'accident' in row.lower():
        incident_type = 'accident'

    # Extract number of cars/vehicles if specified
    numbers = re.findall(r'(\d+)\s*(cars|vehicles)', row, re.IGNORECASE)
    if numbers:
        # If numbers are found, take the first one as the count (assuming one type of vehicle is mentioned)
        num_vehicles_involved = int(numbers[0][0])
    elif 'car' in row.lower() or 'vehicle' in row.lower():
        # Assume at least one vehicle is involved if specific numbers aren't mentioned
        num_vehicles_involved = 1

    return pd.Series([incident_type, num_vehicles_involved])

# Apply the function to the DataFrame
df[['Incident_Type', 'Num_Vehicles_Involved']] = df['Display_Name'].apply(extract_info_from_display_name)


# label_encoders = {}
# for column in ['Main_Category', 'Day', 'Incident_Type','Primary_Vehicle', 'Secondary_Vehicle', 'Is_Major_Incident', 'Closure_Type', 'Direction', 'Traffic_Volume','Advice_A','Advice_B','SA3_NAME21','SA4_NAME21','Suburb','Main_Street']:
#     if column in df.columns:
#         le = LabelEncoder()
#         df[column] = le.fit_transform(df[column].astype(str))
#         label_encoders[column] = le

# Extracting information from "Display_Name"

```

```python
import pandas as pd
from datetime import datetime

def extract_month_and_hour(start_time):
    # Check if the input is a string and parse it if necessary
    if isinstance(start_time, str):
        dt = datetime.strptime(start_time, '%d-%m-%Y %H:%M')
    else:
        dt = start_time
    # Extract month and hour
    month = dt.month
    hour = dt.hour
    return pd.Series([month, hour])

# Sample DataFrame for demonstration
# data = {'Start_Time': ['01-01-2023 08:00', '02-02-2023 09:30', '03-03-2023 10:45']}
# df = pd.DataFrame(data)
df['Start_Time'] = pd.to_datetime(df['Start_Time'], format='%d-%m-%Y %H:%M')

# Apply the function to the DataFrame and create new columns for month and hour
df[['Month', 'Hour']] = df['Start_Time'].apply(extract_month_and_hour)

# Drop unnecessary columns
df = df.drop(['Start_Time'], axis=1)

# Display the DataFrame
df
```

```python
threshold = 0.98  # 90% threshold for imbalance
imbalanced_columns = []

for column in df.columns:
    if df[column].dtype == 'int64' or df[column].dtype == 'object':  # Checking for numerical and categorical data
        total_count = df[column].count()
        most_common_value_count = df[column].value_counts().iloc[0]
        if (most_common_value_count / total_count) > threshold:
            imbalanced_columns.append(column)
#             print(df[column].value_counts())

print("Highly imbalanced columns:", imbalanced_columns)
```

```python
df = df.drop(columns=imbalanced_columns)
```

```python
def sanitize_column_names(df):
    sanitized_names = []
    for col in df.columns:
        # Replace or remove special characters as needed, this is a simple replacement example
        sanitized = col.replace('{', '').replace('}', '').replace('[', '').replace(']', '').replace('"', '').replace(':', '')
        sanitized_names.append(sanitized)
    return sanitized_names

# Sanitize column names
sanitized_column_names = sanitize_column_names(df)

# Apply sanitized column names to the DataFrame
df.columns = sanitized_column_names
```

```python
def analyze_data(df):
    """
    Print column names, top 2 most frequent values from each column, and type of each column.

    Parameters:
    df (pandas.DataFrame): The data frame to analyze.
    """
    for column in df.columns:
        print(f"Column: {column}")
        print(f"Type: {df[column].dtype}")
        print("Top 20 most frequent values:")
        print(df[column].value_counts().head(20))
        print("\n")

analyze_data(df)
```
